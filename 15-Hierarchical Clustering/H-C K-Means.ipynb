# Hierarchical Clustering 


##### Adımlar:

* Her bir data pointi bir cluster olarak seç.
* En yakın iki data pointi alıp bi cluter yap
* en yakın iki cluterı alıp bunları bir cluster yap
* Repeat 3


##### Distance

Eclidean Distance: İki nokta arası mesafe 
1. En yakın iki noktaya göre 
2. En uzak iki noktaya göre
3. Meane göre
4. Centroide göre
 
 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#class 1
x1= np.random.normal(25,5,100)
y1= np.random.normal(25,5,100)

#class2 
x2= np.random.normal(55,5,100)
y2= np.random.normal(60,5,100)

#class3
x3= np.random.normal(55,5,100)
y3= np.random.normal(15,5,100)
```

```python
#x1 x2 ve x3'ü birleştirelim

x= np.concatenate((x1,x2,x2), axis=0)
y= np.concatenate((y1,y2,y3), axis=0)

dictionary= {"x" : x, "y": y}
```

```python
data= pd.DataFrame(dictionary)
```

```python
plt.scatter(x,y, color= "g", alpha=0.3)
plt.show()
```

```python
plt.scatter(x1,y1, color="g", alpha= 0.3, label= "X1-Y1")
plt.scatter(x2,y2, color= "r", alpha= 0.3, label= "X2-Y2")
plt.scatter(x3,y3, color="b", alpha= 0.3, label= "X3-Y3")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
```

# Dendogram Grafiği

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
merg = linkage(data, method="ward") #cluster içindeki yayılımları minimaze etmek
dendrogram(merg,leaf_rotation= 90)
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
h_g= AgglomerativeClustering(n_clusters=3, affinity="euclidean",linkage="ward")
cluster= h_g.fit_predict(data)

```

```python
data["label"]= cluster
```

```python
plt.scatter(data.x[data.label==0],data.y[data.label==0],color="r",label="0")
plt.scatter(data.x[data.label==1],data.y[data.label==1],color="g",label="1")
plt.scatter(data.x[data.label==2],data.y[data.label==0],color="b",label="2")
plt.legend()
plt.show()
```
